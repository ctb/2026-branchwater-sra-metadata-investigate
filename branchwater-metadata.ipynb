```python
df = pl.read_parquet('../2025-workflow-core99/20241128-metadata.parquet')
```

```python
df2 = df.select(['acc', 'organism'])
df2
```

```python
df3 = df.select('organism')
```

```python
df3['organism'].value_counts().sort('count', descending=True)
```

```python
start_keys = df3['organism'].value_counts().sort('count', descending=True)['organism'].to_list()
d = {}
for k in start_keys:
    d[k] = k

d['metagenome'] = 'foo'
d['gut metagenome'] = 'foo'
d['human metagenome'] = 'useless'
d['feces metagenome'] = 'foo'
d['bacterium'] = 'useless'
d['unidentified'] = 'foo'
d['oral metagenome'] = 'useless'
d['skin metagenome'] = 'useless'
d['vaginal metagenome'] = 'useless'

d['Severe acute respiratory syndrome coronavirus 2'] = 'useless'

d['human gut metagenome'] = 'ignore'

d['Homo sapiens'] = 'ignore'
d['Mus musculus'] = 'mouse'


d['human skin metageome'] = 'ignore'
d['human oral metagenome'] = 'ignore'
d['human skin metagenome'] = 'ignore'

d['human vaginal metagenome'] = 'ignore'
d['human nasopharyngeal metagenome'] = 'ignore'
d['human blood metagenome'] = 'ignore'
d['human feces metagenome'] = 'ignore'
d['human lung metagenome'] = 'ignore'


d['mouse gut metagenome'] = 'mouse'
d['mouse metagenome'] = 'mouse'

d['respiratory tract metagenome'] = 'ignore'
#d['pig gut metagenome'] = 'ignore'
d['viral metagenome'] = 'ignore'
```

```python
replace_df = pl.DataFrame(dict(organism=d.keys(), org2=d.values()))
replace_df
```

```python
df3 = df2.join(replace_df, on='organism', how='left')
df3
```

```python
with pl.Config(tbl_rows=-1):
    print(df3['org2'].value_counts().sort('count', descending=True).filter(
        (pl.col("count") > 1000))
    )
```

```python
biglist_sum = df3['org2'].value_counts().sort('count', descending=True).filter(
    pl.col("count") > 3000
)['count'].sum()
```

```python
print(biglist_sum / len(df3))
```

```python
dogs = ['canine metagenome', 'canine gut metagenome']
cats = ['Felis catus', 'feline metagenome']
both = dogs + cats

xx_df = df3.filter(pl.col('organism').is_in(both))
print(len(xx_df))
print(xx_df.filter(pl.col('organism').is_in(cats)))
xx_df['organism'].value_counts()
```

```python
cat_df = xx_df.filter(pl.col('organism').is_in(cats))
accs = cat_df['acc'].to_list()
open("/tmp/cat-list.txt", "w").write("\n".join(accs))
```

```python
cat_df = xx_df.filter(pl.col('organism').is_in(dogs))
accs = cat_df['acc'].to_list()
open("/tmp/dog-list.txt", "w").write("\n".join(accs))
```

```python
from collections import defaultdict
sums = defaultdict(int)

xxx = df3['org2'].value_counts().sort('count', descending=True).iter_rows(named=True)
for row in xxx:
    for word in 'bovine', 'sheep', 'pig', 'Bos taurus', 'goat', 'mouse':
        total = 0
        if word in row['org2']:
            sums[word] += row["count"]

for k, v in sorted(sums.items(), key=lambda x: x[1]):
    print(k, v)
```
